```
f, ax = plt.subplots(nrows=2,ncols=2)
x=np.arange(100)
y=np.random.random(100)
values=np.random.randint(0,50,20)
ax[0,0].plot(x,y)
ax[0,1].scatter(x,y)
ax[1,0].hist(values)
ax[1,1].barh(x,np.abs(y))
```

```
fig,ax=plt.subplots()
x=np.arange(25)
y=np.sin(x)
ax.plot(x,y,color="green" ,linestyle="--",label="sin wave")
x1=np.arange(25)
y1=np.cos(x1)
ax.plot(x1,y1,color="red",linestyle=":",label="cos wave")
ax.legend(loc=4) 
ax.set_xlabel("X axis")  
ax.set_ylabel("y axis") 
ax.set_title("Sin and cos wave") 
```

```
x=np.arange(0,10,2)
y=np.array([11,12,13,17,19])
fig,ax=plt.subplots()
ax.plot(x,y)
for i in zip(x,y):
  ax.annotate(s=str(i)   ,xy=i)
```

```
x=np.arange(0,10,2)
y1=np.sin(x)
y2=np.cos(x)
f,(ax1,ax2)=plt.subplots(nrows=1,ncols=2)
ax1.plot(x,y1)
ax2.plot(x,y2)
```

```
f.savefig('fig.png')
```
